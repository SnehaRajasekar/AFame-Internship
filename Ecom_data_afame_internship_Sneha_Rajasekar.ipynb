```python
ecom_df.drop('Postal Code',axis = 1, inplace = True)
```

**Dropping the Postal Code column as it has more than 90% missing data and it cannot be imputed.**

```python
Total_sales = ecom_df['Sales'].sum()
print('Total Sales amount is ',Total_sales)
```

```python
# Extract year and month from 'Order Date'
ecom_df['Year'] = ecom_df['Order Date'].dt.year
ecom_df['Month'] = ecom_df['Order Date'].dt.month

# Group by year and month, and calculate total sales for each month
monthly_sales = ecom_df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Plot the sales trend over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str), monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### The monthly sales trend shows an increasing trend over the years.

### Top 10 Best selling products

```python
# Group by product and calculate total sales for each product
product_sales = ecom_df.groupby('Product Name')['Sales'].sum().reset_index()

# Sort products by total sales in descending order
product_sales = product_sales.sort_values(by='Sales', ascending=False)

# Select top 10 best-selling products
top_products = product_sales.head(10) 

# Plot the best-selling products
plt.figure(figsize=(10, 6))
plt.barh(top_products['Product Name'], top_products['Sales'], color='skyblue')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.title('Top 10 Best-Selling Products')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
