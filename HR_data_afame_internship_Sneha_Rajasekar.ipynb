**Getting the unique value counts of object data type columns to check for presence of special characters in them.**

```python
# Get object data type columns
object_columns = hr_df.select_dtypes(include=['object']).columns

# Dictionary to store unique value counts for each object column
unique_value_counts = {}

# Calculate unique value counts for each object column
for col in object_columns:
    unique_values = hr_df[col].value_counts()
    unique_value_counts[col] = unique_values

# Print unique value counts for each object column
for col, counts in unique_value_counts.items():
    print(f"Unique value counts for column '{col}':")
    print(counts)
    print()  
```
